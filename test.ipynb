```python
def find_nearest_node(nodes, point):
    nearest_node = None
    nearest_dist = float('inf')
    for node in nodes:
        dist = haversine(node[1], node[0], point[1], point[0])
        if dist < nearest_dist:
            nearest_dist = dist
            nearest_node = node
    return nearest_node
```
