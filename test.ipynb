### Load Modules

```python
from src.pathfinder.utils import *
from src.pathfinder.services import *
from src.pathfinder.visualization import plot_area
```

### Get Area Data

```python
place_names = [
    "Shibuya, Japan", 
    "Shinjuku, Japan",
    "Chiyoda, Japan",
    "Minato, Japan",
    "Chuo, Japan",
    "Taito, Japan",
    "Bunkyo, Japan",
    ]

polygon = get_merged_polygon(place_names)

nodes, edges = get_osm_data_from_poly(polygon, "drive")
edges = clean_edges(edges)
```

### Specify the start and end points

```python
start_node = (35.714796852268535, 139.79697814453934) # Senso-ji
end_node = (35.659711301425666, 139.70075021533643) # Shibuya Crossing

start_node = find_nearest_node(nodes, start_node)
end_node = find_nearest_node(nodes, end_node)

# Ensure start_node and end_node are in the same format as the graph nodes
start_node_coords = (start_node['y'], start_node['x'])
end_node_coords = (end_node['y'], end_node['x'])
```

### Initialize the graph and get the shortest path

```python
G = init_graph(edges)
path = get_path(G, start_node_coords, end_node_coords)
```

### Visualize the data

```python
plot_area(
    # nodes=nodes,
    edges=edges,
    start_node=start_node, 
    end_node=end_node,
    path=path,
    title="A* Pathfinding from Senso-ji to Shibuya Crossing"
)
```
